## Preprocessing

##### Remove years 2000-2004 from Producer Price of agricultural products index

```python
# Extract the year from the 'month' column
ppag['year'] = ppag['Month'].str.extract(r'(\d{4})')

# Convert the 'year' column to numeric
ppag['year'] = pd.to_numeric(ppag['year'])

# Filter rows with years 2000 to 2004
ppag = ppag[(ppag['year'] < 2000) | (ppag['year'] > 2004)]

# Drop the 'year' column if you don't need it anymore
ppag = ppag.drop('year', axis=1)

# Reset the index of the ppag DataFrame to start at 0 again
ppag.reset_index(drop=True, inplace=True)
```

```python
# Drop years before 2005
gdp = gdp[gdp['year'] >= 2005]
fpi = fpi[fpi['Year'] >= 2005]

# Drop values adjusted for inflation assuming other datasets are not adjusted for inflation either
fpi = fpi.drop("Real", axis=1)

# Rename 'Nominal' to FPI
fpi.rename(columns={'Nominal': 'FPI'}, inplace=True)

# Rename 'gdp' to GDP
gdp.rename(columns={'gdp': 'GDP'}, inplace=True)

# Create a shared 'Date' column for all DataFrames for merging

fpi['Date'] = pd.to_datetime(fpi['Year'], format='%Y')
gdp['Date'] = pd.to_datetime(gdp['year'], format='%Y')
df['Date'] = pd.to_datetime(df['Month'], format='%YM%m')

# Set Date as index
fpi.set_index('Date', inplace=True)
gdp.set_index('Date', inplace=True)
df.set_index('Date', inplace=True)

# Remove unnecessary column 'Year'
gdp = gdp.drop('year', axis=1)
fpi = fpi.drop('Year', axis=1)
```

```python
# Merge df and fpi based on the 'Date' index
merged_df = pd.merge(df, fpi, left_index=True, right_index=True, how='outer')

# Merge the resulting DataFrame with gdp based on the 'Date' index
final_merged_df = pd.merge(merged_df, gdp, left_index=True, right_index=True, how='outer')
print(final_merged_df)
```

```python
# Idk why there is a year column again but lets remove year and month because we have date
df = final_merged_df
df = df.drop(['Month', 'Year'], axis=1)
print(df)
```

```python
# Fill empty values with interpolation

df['GDP'] = df['GDP'].interpolate(method='linear')
df['FPI'] = df['FPI'].interpolate(method='linear')
print(df)
```

## Modeling

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt

# Define the split point (e.g., 80% train, 20% test)
split_point = int(0.8 * len(df))

# Split the data into training and test sets without shuffling
train_data = df.iloc[:split_point]
test_data = df.iloc[split_point:]

# Separate features and target variables
X_train = train_data[['CPI', 'CPIE', 'PPAG', 'GDP', 'FPI']]
y_train = train_data['CPIF']

X_test = test_data[['CPI', 'CPIE', 'PPAG', 'GDP', 'FPI']]
y_test = test_data['CPIF']

# Train a Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

# Calculate R^2 (variance score)
r2 = r2_score(y_test, y_pred)

# Calculate variance
variance = np.var(y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
print(f"Variance Score (R^2): {r2:.2f}")
print(f"Variance: {variance:.2f}")

# Visualize the predictions
plt.figure(figsize=(10, 6))
plt.plot(X_test.index, y_test, label="Actual CPIF")
plt.plot(X_test.index, y_pred, label="Predicted CPIF")
plt.xlabel("Date")
plt.ylabel("CPIF")
plt.legend()
plt.title("CPIF Prediction vs. Actual")
plt.show()
```
